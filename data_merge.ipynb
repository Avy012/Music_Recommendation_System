```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/4-2/project/mpd_spotify/userdata/Music Info.csv')
user = pd.read_csv('/content/drive/MyDrive/4-2/project/mpd_spotify/userdata/User Listening History.csv')
```

```python
merged = pd.merge(
    df,        # 왼쪽 테이블 (플레이리스트 트랙)
    user,        # 오른쪽 테이블 (메타 정보)
    on='track_id',
    how='inner',# 공통으로 존재하는 트랙만
    suffixes=('_pl', '_meta')
)
merged.reset_index(drop=True, inplace=True)
merged.drop_duplicates()
```

```python
extract = merged[['spotify_id','name','artist','user_id','playcount','duration_ms']]
#extract['spotify_id'] = 'spotify:track:' + extract['spotify_id']
users = extract.rename(columns={'spotify_id': 'track_uri'})
users.info()
```

```python
lyrics = pd.read_csv('/content/drive/MyDrive/4-2/project/mpd_spotify/lyrics/songs_with_attributes_and_lyrics.csv')
```

```python
lyrics.rename(columns={'id': 'track_uri'}, inplace=True)

merged2 = pd.merge(
    users,        # 왼쪽 테이블 (플레이리스트 트랙)
    lyrics,        # 오른쪽 테이블 (메타 정보)
    on='track_uri',
    how='inner',# 공통으로 존재하는 트랙만
    suffixes=('_pl', '_meta')
)
merged2.drop_duplicates().reset_index(drop=True, inplace=True)
```

```python
merged2.rename(columns={'name_pl': 'track_name','duration_ms_pl':'duration_ms'}, inplace=True)
merged2.drop(['name_meta','album_name','artists','duration_ms_meta'], axis=1, inplace=True)
merged2
```

```python
# CSV로 저장
output_path = '/content/drive/MyDrive/4-2/project/mpd_spotify/user_track2.csv'
merged2.to_csv(output_path, index=False, encoding='utf-8-sig')
print(f"Saved to {output_path}")
```
